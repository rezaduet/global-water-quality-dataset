```python
import pandas as pd
from googletrans import Translator
# Initialize the Translator
try:
    translator = Translator()
except Exception as init_error:
    print(f"Error initializing translator: {init_error}")

# Function to translate text
def translate_text(text):
    if isinstance(text, str):  # Check if the input is a string
        try:
            # Translate from Chinese to English
            translated = translator.translate(text, src='zh-cn', dest='en')
            return translated.text
        except Exception as e:
            print(f"Error occurred while translating '{text}': {e}")
            return text  # Return original text in case of an error
    return text  # Return original value if it's not a string

# Translate column names
translated_columns = [translate_text(col) for col in df.columns]

# Apply translation to all elements in the DataFrame
df_translated = df.applymap(translate_text)  # applymap for element-wise translation

# Assign the translated column names back to the DataFrame
df_translated.columns = translated_columns

# Display the updated DataFrame with translations
df_translated.head()
```

```python
#Run Only if date is unformatted
# Function to clean and convert dates
def clean_and_convert(date_str):
    try:
        date_str = date_str.replace('020', '20')  # Handle the '020' issue
        date = pd.to_datetime(date_str, format='%Y-%m-%d').strftime('%d-%m-%Y')
        return date
    except Exception as e:
        return None  # Handle any other issues with a missing or placeholder value

# Apply the function to the 'parameterSamplingPeriod' column
df_translated['date'] = df_translated['date'].apply(clean_and_convert)

# Print the modified DataFrame
df_translated.head()
```

# WQI Measurement and Classification

Check and convert unformatted data to numeric

```python
# Clean the dataset by removing symbol and converting to float, excluding the first two columns
def clean_column_values(df):
    for column in df.columns[2:]:  # Start from the third column onward
        # Apply conversion for each value in the column
        df[column] = df[column].apply(lambda x: float(x[1:]) if isinstance(x, str) and x.startswith('>') else pd.to_numeric(x, errors='coerce'))
    return df

# Clean the DataFrame
df = clean_column_values(df)

#Run this cell again if error encountered in the next cell
```

Calculating CCME Water Quality Indexing

```python
import numpy as np
#If this cell show error, rerun the previous cell
df = new_df
# Get the Ireland WQI standards - Si value
ireland_wqi_standards = {
  'Ammonia (mg/l)': 0.5,
  'Biochemical Oxygen Demand (mg/l)': 6,
  'Dissolved Oxygen (mg/l)': 14,
  'Orthophosphate (mg/l)': 0.05,
  'pH (ph units)': 8.5,
  'Temperature (cel)': 25,
  'Nitrogen (mg/l)': 20,
  'Nitrate (mg/l)': 45
}

ex1 = 0
ex2 = 0
failed_variables = []
columns = len(df.columns)
#print(columns)
WQI = []
for i in range(len(df)):
  fv = 0
  excurstion = 0
  for j in range(columns):
    param_index = j  # Specify the index of the parameter you want to check (0-based index)
    param_list = list(ireland_wqi_standards.items())
    # print("The Pram Liss >>>>>>>>>>>>>>>>>", param_list)
    if param_index < len(param_list):
        parameter, standard = param_list[param_index]

        # print("Parameters>>>>>>>>", parameter, "Standards<<<<<<<<<<", standard)
        if parameter in df:
            value = df[parameter].values[i]  # 0 means the first row value
            #print("Param=",parameter, "Value=", value, "Std:", standard)
            # print("The Standrad Value is1:", standard[0])
            # print("The Standrad Value 2 is:", standard[1])
            # print("observed Value:>>>>>>", value, "| Standard Value>>>>>", standard)
            if value > standard: #Checking failed values
              fv += 1
              excurstion += (value/standard)-1

  # print("Excursion Value:", excurstion, "Excurtion 1 & 2", ex1, ex2, "Summation", ex1+ex2)
  # print(fv)
  # break
  #Calculatig F1

  F1 = (fv/columns)*100
  F2 = F1 #Since we are calculating a single entry data (a row) instead of a data of time span (a week/month/year/ sepcific time span)

  nse = excurstion/columns
  F3 = nse/(0.01*nse + 0.01)
  wqi_sum = 100-(np.sqrt(F1*F1+F2*F2+ F3*F3)/1.732)
  WQI.append(wqi_sum)
print("WQI value is: ", WQI)


#print(failed_variables)

#Calculating F1
```

```python
final_df = df
final_df['CCME_Values'] = WQI
```

```python
#Classifying Water Quality Iindex Values
result_CCME = []
j=0
for i in range(len(WQI)):
  if WQI[i]<=100 and WQI[i]>=95:
    result_CCME.append('Excellent')
  elif WQI[i]<95 and WQI[i]>=80:
    result_CCME.append("Good")
  elif WQI[i]<80 and WQI[i]>= 65:
    result_CCME.append("Fair")
  elif WQI[i]<65 and WQI[i]>= 45:
    result_CCME.append("Marginal")
  elif WQI[i]<45 and WQI[i]>= 0:
    result_CCME.append("Poor")
  else:
    result_CCME.append('Unsuitable')
    #print(WQI[i])

print(result_CCME)
```
